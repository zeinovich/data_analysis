```python
import pickle
import numpy as np
import pandas as pd
```

```python
with open('TrainData/merged_data.csv', 'rb') as f:
    data = pd.read_csv(f)

data.isna().sum()
```

```python
data.TypeOfCollission = data.TypeOfCollission.fillna('NAN')
data.PropertyDamage = data.PropertyDamage.fillna('NAN')
data.PoliceReport = data.PoliceReport.fillna('NAN')
data.isna().sum()
```

```python
data = data.dropna()
data.isna().sum()
```

```python
datatypes = pd.DataFrame(data.dtypes, columns=['dtype'])
datatypes
```

```python
data[datatypes[datatypes['dtype'] == 'object'].index] = data[datatypes[datatypes['dtype'] == 'object'].index].astype('category')
datatypes = pd.DataFrame(data.dtypes, columns=['dtype'])
datatypes
```

```python
from sklearn.preprocessing import MinMaxScaler

data['DateOfIncident'] = pd.to_datetime(data['DateOfIncident'].astype(str))
data['DateOfPolicyCoverage'] = pd.to_datetime(data['DateOfPolicyCoverage'].astype(str))
data['VehicleYOM'] = pd.to_datetime(data['VehicleYOM'].astype(str))
data['VehicleAge'] = data['DateOfIncident'].dt.year - data['VehicleYOM'].dt.year
data['PolicyAge'] = data['DateOfIncident'].dt.year - data['DateOfPolicyCoverage'].dt.year
datatypes = pd.DataFrame(data.dtypes, columns=['dtype'], index=data.columns)
num_cols = datatypes[datatypes['dtype'] != 'category'].index
num_cols = num_cols.tolist()
num_cols.remove('ReportedFraud')
num_cols.remove('Witnesses')
num_cols.remove('NumberOfVehicles')
cat_cols = datatypes[datatypes['dtype'] == 'category'].index
cat_cols = cat_cols.tolist()
cat_cols.extend(['Witnesses', 'NumberOfVehicles'])
datatypes = pd.DataFrame(data.dtypes, columns=['dtype'], index=data.columns)
dates = datatypes[datatypes['dtype'] == 'datetime64[ns]'].index
dates = dates.tolist()
data[dates] = data[dates] - pd.Timestamp("1980-01-01")
data[dates] = data[dates].astype('int64')
scaler = MinMaxScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
train = pd.get_dummies(data, columns=cat_cols)
```
