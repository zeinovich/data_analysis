```python
df = pd.merge(target_df, claim_df, on='CustomerID', how='left')
df = pd.merge(df, policy_df, on='CustomerID', how='left')
df = pd.merge(df, vehicle, on='CustomerID', how='left')
df = pd.merge(df, demogr_df, on='CustomerID', how='left')
df.head()
```

```python
data = df.copy()
data['CustomerID'] = data['CustomerID'].apply(lambda x: x.replace('Cust', ''))
data['CustomerID'] = data['CustomerID'].astype(int)
data.head()
```

```python
data['IncidentState'] = data['IncidentState'].apply(lambda x: x.replace('State', ''))
data['IncidentCity'] = data['IncidentCity'].apply(lambda x: x.replace('City', ''))
data['IncidentAddress'] = data['IncidentAddress'].apply(lambda x: x.replace('Location ', ''))
data['InsurancePolicyState'] = data['InsurancePolicyState'].apply(lambda x: x.replace('State', ''))
data.head()
```

```python
# Expand PolicyCombinedSingleLimit
data['PolicyCombinedLimit'] = data['Policy_CombinedSingleLimit'].apply(lambda x: x.split('/')[0]).astype(int)
data['PolicySingleLimit'] = data['Policy_CombinedSingleLimit'].apply(lambda x: x.split('/')[1]).astype(int)
data = data.drop('Policy_CombinedSingleLimit', axis=1)
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
datatypes = pd.DataFrame(data.dtypes, columns=['dtype'], index=data.columns)
datatypes
```

```python
cat_cols = datatypes[datatypes['dtype'] == 'object'].index.tolist()
cat_cols.extend(['NumberOfVehicles', 'BodilyInjuries', 'Witnesses', 'InsuredZipCode'])
cat_cols.remove('ReportedFraud')
cat_cols.remove('PoliceReport')
cat_cols
```

```python
num_cols = datatypes[datatypes['dtype'] != 'object'].index.tolist()
num_cols = [col for col in num_cols if col not in cat_cols]
num_cols
```

```python
from sklearn.impute import SimpleImputer

cat_imputer = SimpleImputer(strategy='most_frequent')
num_imputer = SimpleImputer(strategy='mean')

cat_imputer.fit(data[cat_cols])
num_imputer.fit(data[num_cols])

data_wo_nan = data.copy()
data_wo_nan['PoliceReport'] = data_wo_nan['PoliceReport'].fillna('NA')
data_wo_nan[cat_cols] = cat_imputer.transform(data[cat_cols])
data_wo_nan[num_cols] = num_imputer.transform(data[num_cols])

data_wo_nan.isna().sum()
```

```python
import pickle

with open('models/cat_imputer.pkl', 'wb') as f:
    pickle.dump(cat_imputer, f)

with open('models/num_imputer.pkl', 'wb') as f:
    pickle.dump(num_imputer, f)
```

```python
data_wo_nan[cat_cols] = data_wo_nan[cat_cols].astype('category')

data_wo_nan.info()
```

```python
with open('TrainData/train_data_prepr.pkl', 'wb') as f:
    data_wo_nan.to_pickle(f)
```
