```python
import pandas as pd

with open('data/2015.csv', 'r') as f:
    data_2015 = pd.read_csv(f, usecols=['Country', 'Happiness Score', 'Economy (GDP per Capita)'])

with open('data/2016.csv', 'r') as f:
    data_2016 = pd.read_csv(f, usecols=['Country', 'Happiness Score', 'Economy (GDP per Capita)'])

with open('data/2017.csv', 'r') as f:
    data_2017 = pd.read_csv(f, usecols=['Country', 'Happiness.Score', 'Economy..GDP.per.Capita.'])

with open('data/2018.csv', 'r') as f:
    data_2018 = pd.read_csv(f, usecols=['Country or region', 'Score',  'GDP per capita'])

with open('data/2019.csv', 'r') as f:
    data_2019 = pd.read_csv(f, usecols=['Country or region', 'Score', 'GDP per capita'])

#Rename columns
data_2015.rename(columns={'Country': 'country', 'Happiness Score': 'score_2015', 'Economy (GDP per Capita)': 'economy_2015'}, inplace=True)
data_2016.rename(columns={'Country': 'country', 'Happiness Score': 'score_2016', 'Economy (GDP per Capita)': 'economy_2016'}, inplace=True)
data_2017.rename(columns={'Country': 'country', 'Happiness.Score': 'score_2017', 'Economy..GDP.per.Capita.': 'economy_2017'}, inplace=True)
data_2018.rename(columns={'Country or region': 'country', 'Score': 'score_2018', 'GDP per capita': 'economy_2018'}, inplace=True)
data_2019.rename(columns={'Country or region': 'country', 'Score': 'score_2019', 'GDP per capita': 'economy_2019'}, inplace=True)
#Merge data
data = data_2015.merge(data_2016, on='country', how='outer')
data = data.merge(data_2017, on='country', how='outer')
data = data.merge(data_2018, on='country', how='outer')
data = data.merge(data_2019, on='country', how='outer')

data.head()
```

Now we'll use linear regression to predict economy and scores for 2020, 2021, 2022

```python
score = data[['country', 'score_2015', 'score_2016', 'score_2017', 'score_2018', 'score_2019']]
economy = data[['country', 'economy_2015', 'economy_2016', 'economy_2017', 'economy_2018', 'economy_2019']]
score.head()
```

```python
score.isnull().sum()
```

```python
score = score.dropna()
economy = economy.dropna()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
# Create model
model = LinearRegression()
X = np.array([[2015, 2016, 2017, 2018, 2019]])

score['score_2020'] = np.zeros(len(score))
score['score_2021'] = np.zeros(len(score))
score['score_2022'] = np.zeros(len(score))

for i in range(len(score)):
    y = np.array([score.iloc[i, 1:6]])
    model.fit(X.T, y.T)
    score.iloc[i, 6] = model.predict(np.array([[2020]]))[0][0]
    score.iloc[i, 7] = model.predict(np.array([[2021]]))[0][0]
    score.iloc[i, 8] = model.predict(np.array([[2022]]))[0][0]

score.head()
```

```python
model = LinearRegression()
X = np.array([[2015, 2016, 2017, 2018, 2019]])

economy['economy_2020'] = np.zeros(len(economy))
economy['economy_2021'] = np.zeros(len(economy))
economy['economy_2022'] = np.zeros(len(economy))

for i in range(len(economy)):
    y = np.array([economy.iloc[i, 1:6]])
    model.fit(X.T, y.T)
    economy.iloc[i, 6] = model.predict(np.array([[2020]]))[0][0]
    economy.iloc[i, 7] = model.predict(np.array([[2021]]))[0][0]
    economy.iloc[i, 8] = model.predict(np.array([[2022]]))[0][0]

economy.head()
```

```python
score = score[['country', 'score_2020', 'score_2021', 'score_2022']]
economy = economy[['country', 'economy_2020', 'economy_2021', 'economy_2022']]
score.head()
```

```python
data = score.merge(economy, on='country', how='outer')
data.head()
```

```python
with open('iso_short.csv', 'r') as f:
    iso = pd.read_csv(f)

iso
```

```python
def to_iso(country):
    code = iso.loc[iso['name'] == country, 'code']
    return code.values[0] if not code.empty else country

data.loc[:, 'country'] = data.loc[:, 'country'].apply(to_iso)
data.head()
```

```python
with open('scores.csv', 'wb') as f:
    data.to_csv(f, index=False)
```
